# Task 3: Geospatial Analysis 

```python
import plotly.express as px

fig = px.scatter_mapbox(data2, lat="Latitude", lon="Longitude", hover_name="Restaurant Name", hover_data=["City", "Cuisines"])
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0, "t":0, "l":0,"b":0})
fig.show()
```

```python
country_res_dis = data2["Country Code"].value_counts()
print(country_res_dis)
```

```python
city_res_dis = data2["City"].value_counts()
print(city_res_dis)
```

```python
fig = px.scatter(data2, x="Longitude", y = "Latitude", color= "Aggregate rating", hover_name= "Restaurant Name", color_continuous_scale="viridis")
fig.update_layout(title="Scatter Ployt vs Aggregate Rating", xaxis_title="Longitude", yaxis_title="latitude")
fig.show()
```
